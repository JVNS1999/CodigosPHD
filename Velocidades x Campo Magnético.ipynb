```python
import numpy as np
import matplotlib.pyplot as plt
import scienceplots
plt.style.use(['science','notebook','grid'])
```

```python
# Dados dos pontos
campo_magnetico = np.array([1.0, 1.0])  # Ajustando para apenas 1.0 mT
velocidade_1 = 440  # Velocidade para 1.0 mT (triângulo azul)
velocidade_2 = 378  # Velocidade para 1.0 mT (bolinha vermelha)

# Criando o gráfico
plt.figure(figsize=(10, 6))

# Adicionando os pontos com os respectivos marcadores
plt.scatter([1.0], [velocidade_1], color='blue', marker='^', s=100)  # Triângulo azul
plt.scatter([1.0], [velocidade_2], color='red', marker='o', s=100, facecolors='none')  # Bolinha vermelha

# Se você adicionar mais pontos, pode conectar eles por uma linha pontilhada
# Exemplo com mais pontos fictícios
campo_magnetico_adicional = np.array([1.0, 1.0, 1.0])
velocidade_adicional = np.array([440, 380, 370])

# Adicionando a linha pontilhada entre os pontos
plt.plot(campo_magnetico_adicional, velocidade_adicional, linestyle=':', color='black')  # Linha pontilhada

# Títulos e rótulos com fontes maiores
plt.title("Gráfico de Velocidade x Campo Magnético", fontsize=16)
plt.xlabel("Campo Magnético (mT)", fontsize=14)
plt.ylabel("Velocidade (m/s)", fontsize=14)

# Melhorando o grid e as linhas
plt.grid(True, linestyle='--', alpha=0.7)
plt.axhline(0, color='black',linewidth=1)
plt.axvline(0, color='black',linewidth=1)

# Exibindo o gráfico
plt.show()
```
